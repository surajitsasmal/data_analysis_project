# Rural heathcare of India

In 2019, India's rural health sector faced significant challenges. The government initiated the Ayushman Bharat
program, aiming to provide financial protection and access to quality healthcare for millions of rural Indians through
the Pradhan Mantri Jan Arogya Yojana (PM-JAY) scheme. However, despite these efforts, disparities in healthcare infrastructure
and access persisted, particularly in remote areas.

In 2020, India's rural health sector faced an unprecedented crisis with the outbreak of the COVID-19 pandemic.
The pandemic strained already limited healthcare resources in rural areas, highlighting the urgent need for
infrastructure improvement and healthcare capacity building. The government's response included setting up temporary
healthcare facilities, deploying healthcare workers, and raising awareness about preventive measures. These years underscored
the importance of addressing healthcare inequalities in India's rural regions and the need for sustained investment and
reforms in the rural health sector.


Here we   analyze different Healthcare facilities available to rural India in various States
and Union Territories in 2019 and 2020.
